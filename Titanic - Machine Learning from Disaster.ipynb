```python
plt.figure(figsize=(8, 4))
plt.title('Sex Percentage')
plt.ylabel('Percent')
percent=df.Sex.value_counts()*100/len(df)

# Plot the bars
bars = percent.plot(kind='bar')

# Annotate each bar
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, round(yval, 2), ha='center', va='bottom')

plt.show()
```

```python
# Assuming df is your DataFrame
plt.figure(figsize=(8, 4))
plt.title('Percentage of Survival')
plt.ylabel('Percentage')

# This is your series of percentages
percentages = (df.Survived.value_counts() * 100 / len(df))

# Plot the bars
bars = percentages.plot(kind='bar')

# Annotate each bar
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, round(yval, 2), ha='center', va='bottom')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
plt.figure(figsize=(10, 6))
ax = pd.crosstab(df['Survived'], df['Sex'], normalize='columns').plot(kind='bar')

plt.title('Percentage of Survival by Sex')
plt.ylabel('Percent')

# Adding annotations with percentages
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2.,
            height + 0.02,  # small offset to place text above the bar
            '{:1.2f}%'.format(height*100),
            ha="center")

plt.show()
```
